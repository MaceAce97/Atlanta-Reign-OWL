```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

stats = pd.read_csv('2022-AR-OWL-Stats.csv')
stats
```

```python
players = [stats["Player_Name"]]
team = [stats["Team"]]
wins= [stats["Win_Percentage"]]

print(players)
print(wins)

sum(wins)
```

```python
PlyNam = stats["Player_Name"]
WP = stats["Win_Percentage"]
plt.figure(figsize = (14,6))
plt.bar(PlyNam, WP, color='C3')
plt.xlabel("Player Name", font='monospace', fontsize=16)
plt.ylabel("Win Percentage",font='monospace', fontsize=16)
plt.title("Overall Win Percentage for Atlanta Reign Players", fontsize=20)
```

```python
PlyNam = stats["Player_Name"]
Dif = stats["Differential"]
#If the Dif is in the negative (-) then they have lost more than win, and if it's positive, they've won more than lost during their plays
plt.figure(figsize = (14,6))
plt.bar(PlyNam, Dif, color='r')
plt.xlabel("Player Name", font='monospace', fontsize=16)
plt.ylabel("Differential",font='monospace', fontsize=16)
plt.title("Overall Result Differential for Atlanta Reign Players", fontsize=20)
```
